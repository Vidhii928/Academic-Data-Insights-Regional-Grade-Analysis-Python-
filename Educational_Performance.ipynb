```python
# create a pie chart for data grouped on the basis of grades
import matplotlib.pyplot as plt
grouped_data = df.groupby('Grade').size()
labels = ["Akanshi-1","Ati-Uttam","Prachesta-1","Prachesta-2","Prachesta-3","Uttam"]
values = [1,51,290,117,18,271]
plt.pie(values, labels=labels, autopct="%1.1f%%")
plt.title("Pie Chart for Grade")
plt.show()
# autopct: We use autopct to display the percent value using Python string formatting. For example, autopct='%1.1f%%' means that for each pie wedge, the format string is '1.1f%'.
```

```python
# data grouped on the basis states and grades
df.groupby(['State/UT','Grade']).size()
```

```python
# draw a histogram on the basis of grouped data of states and grades

grouped_data = df.groupby(['State/UT','Grade']).size().unstack()
grouped_data.plot(kind='bar', stacked=True)
plt.xlabel("State/UT")
plt.ylabel("Number of Students")
plt.title("Number of Students in Each State/UT by Grade")
plt.show()
#stack() : stack the prescribed level(s) from column to row.
#unstack() : unstack the prescribed level(s) from row to column.
```
